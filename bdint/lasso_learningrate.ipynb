```python
import pandas as pd
import numpy as np
from bdint.models import CatBoost, RandomForest, LinearRegression
import matplotlib.pyplot as plt
from bdint.data import (
    get_test_df,
    get_train_df, k_fold_validation)

```

```python
import time
learning_rates_log = np.linspace(-7, 1, 30)
times = []
errors = []

for l in learning_rates_log:
    print("LR: ", l)
    train_df = get_train_df()
    test_df = get_test_df()

    timeA = time.time()
    model = LinearRegression(train=train_df, test=test_df, regression_type="Lasso", skewness_threshold=1.0, alpha = 10**l)

    rmse = k_fold_validation(model)
    times.append(time.time()-timeA)
    errors.append(rmse)

```

```python
# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the first data series
color = "tab:red"
ax1.set_xlabel("Learning Rates [log10]")
ax1.set_ylabel("RMSE", color=color)
ax1.plot(learning_rates_log, errors, color=color)
ax1.tick_params(axis="y", labelcolor=color)

# Create a second y-axis
ax2 = ax1.twinx()
color = "tab:blue"
ax2.set_ylabel("Time [s]", color=color)
ax2.plot(learning_rates_log, times, color=color)
ax2.tick_params(axis="y", labelcolor=color)

plt.title("Lasso Regression Learning Rates")

plt.savefig(f'vizualization/lasso_learning_rates.png')
plt.show()
```
